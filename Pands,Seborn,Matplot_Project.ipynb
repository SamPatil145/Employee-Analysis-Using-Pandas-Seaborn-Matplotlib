Import Libraries

PART 2: MATPLOTLIB AND SEBORN (Charts ONLY)

```python
# Q1. Number of employees per department (Bar Chart)

dept_counts = df['Department_Code'].value_counts()
plt.bar(dept_counts.index, dept_counts.values, edgecolor="black", color="pink")
plt.xlabel("Department")
plt.ylabel("Employee Count")
plt.title("Employees per Department")
plt.show()
```

```python
# Q2.Age distribution of employees

plt.hist(df['Age'], edgecolor="black", color="pink")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

```python
# Q3. Average salary by department

avg_sal = df.groupby('Department_Code')['Salary'].mean()
plt.bar(avg_sal.index, avg_sal.values, edgecolor="black", color="pink")
plt.title("Average Salary by Department")
plt.ylabel("Salary")
plt.show()
```

```python
# Q4. Gender distribution (Pie Chart)

colour=['pink','blue']

gender_count = df['Gender'].value_counts()
plt.pie(gender_count.values, labels=gender_count.index, autopct='%1.1f%%', colors=colour)
plt.title("Gender Distribution")
plt.legend(loc='upper right')
plt.show()
```

```python
# Q5. Department vs Salary (Box Plot)

sns.boxplot(x='Department_Code', y='Salary', data=df, palette='rocket')
plt.title("Salary by Department")
plt.show()
```

```python
# Q6. Gender vs Salary

sns.boxplot(x='Gender', y='Salary', data=df, palette='rocket' )
plt.title("Salary by Gender")
plt.show()
```

```python
# Q7.Employment type vs Salary

sns.boxplot(x='Employment_Type', y='Salary', data=df, palette='rocket')
plt.title("Employment Type vs Salary")
plt.show()
```

```python
# Q7. How does average salary change with experience level?

exp_salary = df.groupby('Experience_Years')['Salary'].mean()
plt.plot(exp_salary.index, exp_salary.values, marker='*', mec='black', color='pink' ,mfc='blue')
plt.xlabel("Experience Years")
plt.ylabel("Average Salary")
plt.title("Salary Trend by Experience")
plt.show()
```

```python
# Q8. What is the employment type distribution?

emp_type = df['Employment_Type'].value_counts()

plt.pie(emp_type.values, labels=emp_type.index, autopct='%1.1f%%')
plt.title("Employment Type Distribution")
plt.legend()
plt.show()
```

```python
# Q9. Distribution of performance score?

sns.histplot(df['Performance_Score'], kde=True, color='pink', edgecolor='black')
plt.title("Performance Rating Distribution")
plt.show()
```

```python
# Q10. Average salary change with years of experience?

avg_salary = df.groupby('Experience_Years')['Salary'].mean()
plt.fill_between(avg_salary.index, avg_salary.values, color='pink')
plt.xlabel("Experience Years")
plt.ylabel("Average Salary")
plt.title("Average Salary by Experience (Area Chart)")
plt.show()
```
